```python
import pandas as pd
from tqdm import tqdm

# merge data
source_pd_data = pd.read_csv(src_files[0].absolute(),encoding='gbk',dtype='float32')
source_pd_data.columns = new_col
for file in tqdm(src_files[1:]):
    # 忽略第一行
    d = pd.read_csv(file.absolute(),encoding='gbk',dtype='float32')
    d.columns = new_col
    source_pd_data = pd.concat([source_pd_data,d])
    
print(source_pd_data.isnull().sum())
source_pd_data.to_csv("./source_data.csv",index=False)
source_pd_data.head(30)
    
```
